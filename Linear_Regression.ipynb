```python
file_path = '/content/drive/My Drive/ML/Linear_Regression_Data/Data.csv'
```

```python
# Load the data
data = pd.read_csv(file_path)
data.head()
data.head(data.shape[0])
```

```python
X = data['X'].values  # Independent variable
y = data['y'].values  # Dependent variable
```

```python
# Scatter plot of the data
plt.scatter(X, y)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Scatter Plot of the Data')
plt.show()
```

```python
# Initialize parameters
w0 = 300
w1 = 10
learning_rate = 0.0001
epochs = 10  # Number of iterations for gradient descent

n = float(len(X))  # Number of elements in X

# Perform Gradient Descent
for i in range(epochs):
    y_pred = w0 + w1 * X  # Predicted value of y
    D_w0 = (2/n) * sum(y_pred - y)  # Derivative w.r.t w0
    D_w1 = (2/n) * sum(X * (y_pred - y))  # Derivative w.r.t w1

    w0 = w0 - learning_rate * D_w0  # Update w0
    w1 = w1 - learning_rate * D_w1  # Update w1

    print(f"Epoch {i+1}: w0 = {w0}, w1 = {w1}")
```

```python
predictions = w0 + w1 * X
predictions
```

```python
# Plot the results
plt.scatter(X, y, label='Actual Data')
plt.plot(X, predictions, color='red', label='Fitted Line')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression')
plt.legend()
plt.show()
```
